```python
df['University_Name'] = df['University_Name'].str.replace(r'\d+', '', regex=True).str.strip()
# str.replace(r'\d+', '', regex=True) - removes all digits
# str.strip() - removes leading/trailing whitespace left after removing numbers
# regex=True - tells pandas to treat the pattern as a regular expression
```

```python
df.head(50)
```

```python
df.columns = [col.lower() for col in df.columns]
df.columns
```

```python
df['student_faculty_ratio_calc'] = df['undergrad_students'] + df['postgrad_students'] / df['faculty_count']
```

```python
# Selectivity category (based on acceptance rate)
df['selectivity'] = pd.cut(df['acceptance_rate (%)'], 
                           bins=[0, 20, 40, 60, 100], 
                           labels=['Highly Selective', 'Selective', 'Moderately Selective', 'Less Selective'])
```

```python
# Convert Research_Output to numeric
df['research_output'] = pd.to_numeric(df['research_output'], errors='coerce')

# Now calculate research per faculty
df['research_per_faculty'] = (df['research_output'] / df['faculty_count']).round(2)
```

```python
# Tuition category
df['tuition_category'] = pd.cut(df['tuition_fee_usd'], 
                                bins=[0, 10000, 30000, 50000, float('inf')], 
                                labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df['enrollment_category'] = pd.cut(df['enrollment'], 
                                    bins=[0, 10000, 20000, 30000, float('inf')],
                                    labels=['Low ', 
                                            'Medium ', 
                                            'High ', 
                                            'Very High '])
```

```python
# International students count
df['international_students_count'] = (df['enrollment'] * df['international_students (%)'] / 100).round()
```

```python
#  Average rank (across different ranking systems)
df['average_rank'] = df[['world_rank', 'qs_rank', 'the_rank', 'arwu_rank']].mean(axis=1).round()
```

**Linear Regression Analysis: University Metrics Comparison**

**1. Tuition Fee vs World Rank**
- Relationship: Nearly flat line (very weak/no correlation)

- Interpretation: Expensive universities aren't necessarily highly ranked, and vice versa

**2. Acceptance Rate vs Graduation Rate**
- Relationship: Nearly flat line (very weak/no correlation)
- Pattern: Random scatter across all values (60-97% graduation rates at all acceptance levels)
- Conclusion: Acceptance rate does not predict graduation rate
- Interpretation: Selective universities don't necessarily have higher graduation rates

**3. International Students vs World Rank**
- Relationship: Slightly declining line (weak negative correlation)
- Pattern: Some downward trend but with lots of scatter
- Conclusion: Top-ranked universities have slightly more international students
- Interpretation: Better universities attract slightly more international students, but it's a weak relationship

**4. Enrollment vs Faculty Count**
- Relationship: Strong positive correlation (clear upward trend)
- Pattern: Points follow the regression line closely
- Conclusion: Strong predictor - larger enrollment consistently means more faculty
- Interpretation: Universities maintain relatively consistent student-faculty ratios regardless of size

**Summary:**

| Relationship | Correlation | Pattern | Predictive Power |
|-------------|-------------|---------|------------------|
| Tuition → Rank | Very Weak | Random | None |
| Acceptance → Graduation | Very Weak | Random | None |
| International % → Rank | Weak Negative | Slight trend | Low |
| Enrollment → Faculty | Strong Positive | Clear trend | High |

**Key Findings:**
- Only enrollment and faculty count show a strong, predictable relationship
- University rankings, tuition, acceptance rates, and graduation rates are influenced by many complex factors
- Simple linear relationships don't explain most university metrics
- The enrollment-faculty relationship makes logical sense: more students require more teachers

**Conclusion:**
Most university success metrics cannot be predicted by single variables. Only operational metrics like enrollment and faculty count show strong correlations, as they're directly related by institutional needs.
