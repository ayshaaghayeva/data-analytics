```python
df['University_Name'] = df['University_Name'].str.replace(r'\d+', '', regex=True).str.strip()
# str.replace(r'\d+', '', regex=True) - removes all digits
# str.strip() - removes leading/trailing whitespace left after removing numbers
# regex=True - tells pandas to treat the pattern as a regular expression
```

```python
df.head(50)
```

```python
df.columns = [col.lower() for col in df.columns]
df.columns
```

```python
df['student_faculty_ratio_calc'] = df['undergrad_students'] + df['postgrad_students'] / df['faculty_count']
```

```python
# Selectivity category (based on acceptance rate)
df['selectivity'] = pd.cut(df['acceptance_rate (%)'], 
                           bins=[0, 20, 40, 60, 100], 
                           labels=['Highly Selective', 'Selective', 'Moderately Selective', 'Less Selective'])
```

```python
# Convert Research_Output to numeric
df['research_output'] = pd.to_numeric(df['research_output'], errors='coerce')

# Now calculate research per faculty
df['research_per_faculty'] = (df['research_output'] / df['faculty_count']).round(2)
```

```python
# Tuition category
df['tuition_category'] = pd.cut(df['tuition_fee_usd'], 
                                bins=[0, 10000, 30000, 50000, float('inf')], 
                                labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df['enrollment_category'] = pd.cut(df['enrollment'], 
                                    bins=[0, 10000, 20000, 30000, float('inf')],
                                    labels=['Low ', 
                                            'Medium ', 
                                            'High ', 
                                            'Very High '])
```

```python
# International students count
df['international_students_count'] = (df['enrollment'] * df['international_students (%)'] / 100).round()
```

```python
#  Average rank (across different ranking systems)
df['average_rank'] = df[['world_rank', 'qs_rank', 'the_rank', 'arwu_rank']].mean(axis=1).round()
```

```python

# Calculate correlation
x = df['world_rank']
y = df['acceptance_rate (%)']

correlation = x.corr(y)
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f"Correlation coefficient: {correlation:.4f}")
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4f}")

# Add stats to plot
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='world_rank', y='acceptance_rate (%)', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Linear Regression: World Rank vs Acceptance Rate\nR² = {r_value**2:.3f}, Correlation = {correlation:.3f}')
plt.xlabel('World Rank')
plt.ylabel('Acceptance Rate (%)')
plt.show()
```

**Linear Regression: World Rank vs Acceptance Rate**

**Key Findings:**

**1. Very weak positive correlation**
- The regression line is nearly flat
- Almost no relationship between world rank and acceptance rate

**2. High variability**
- Universities at ANY world rank can have acceptance rates from 0% to 80%
- World rank doesn't strongly predict acceptance rate

**3. Slight upward trend**
- Lower-ranked universities (higher numbers) have *slightly* higher acceptance rates
- But the effect is minimal

**Conclusion:**
World rank and acceptance rate are **weakly related**. Top-ranked universities don't necessarily have low acceptance rates, and lower-ranked universities don't always have high acceptance rates.

**This means:** Acceptance rate depends on many other factors beyond world rank (like country, program type, university mission, etc.)

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='tuition_fee_usd', y='world_rank', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'green'})
plt.title('Linear Regression: Tuition Fee vs World Rank')
plt.xlabel('Tuition Fee (USD)')
plt.ylabel('World Rank')
plt.show()
```

```python
# Calculate correlation and regression
x = df['enrollment']
y = df['faculty_count']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='enrollment', y='faculty_count', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Enrollment vs Faculty Count\nR² = {r_value**2:.3f}, p-value = {p_value:.4f}')
plt.xlabel('Enrollment')
plt.ylabel('Faculty Count')
plt.show()

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4f}")
```

**Linear Regression: Enrollment vs Faculty Count**

**Key Findings:**
- Strong positive correlation (R² = 0.709)
- Highly significant relationship (p-value < 0.0001)
- 70.9% of faculty count variation is explained by enrollment
- Clear upward trend with some scatter

**Conclusion:** Universities with higher enrollment hire proportionally more faculty to maintain teaching capacity. This is a strong, predictable relationship.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Tuition vs World Rank
sns.regplot(data=df, x='tuition_fee_usd', y='world_rank', ax=axes[0, 0], 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[0, 0].set_title('Tuition Fee vs World Rank')

# Plot 2: Acceptance Rate vs Graduation Rate
sns.regplot(data=df, x='acceptance_rate (%)', y='graduation_rate (%)', ax=axes[0, 1],
            scatter_kws={'alpha':0.5}, line_kws={'color':'green'})
axes[0, 1].set_title('Acceptance Rate vs Graduation Rate')

# Plot 3: International Students vs World Rank
sns.regplot(data=df, x='international_students (%)', y='world_rank', ax=axes[1, 0],
            scatter_kws={'alpha':0.5}, line_kws={'color':'blue'})
axes[1, 0].set_title('International Students vs World Rank')

# Plot 4: Enrollment vs Faculty Count
sns.regplot(data=df, x='enrollment', y='faculty_count', ax=axes[1, 1],
            scatter_kws={'alpha':0.5}, line_kws={'color':'purple'})
axes[1, 1].set_title('Enrollment vs Faculty Count')

plt.tight_layout()
plt.show()
```

**Linear Regression Analysis: University Metrics Comparison**

**1. Tuition Fee vs World Rank**
- Relationship: Nearly flat line (very weak/no correlation)
- Pattern: Random scatter with no clear trend
- Conclusion: Tuition fee does not predict world rank
- Interpretation: Expensive universities aren't necessarily highly ranked, and vice versa

**2. Acceptance Rate vs Graduation Rate**
- Relationship: Nearly flat line (very weak/no correlation)
- Pattern: Random scatter across all values (60-97% graduation rates at all acceptance levels)
- Conclusion: Acceptance rate does not predict graduation rate
- Interpretation: Selective universities don't necessarily have higher graduation rates

**3. International Students vs World Rank**
- Relationship: Slightly declining line (weak negative correlation)
- Pattern: Some downward trend but with lots of scatter
- Conclusion: Top-ranked universities have slightly more international students
- Interpretation: Better universities attract slightly more international students, but it's a weak relationship

**4. Enrollment vs Faculty Count**
- Relationship: Strong positive correlation (clear upward trend)
- Pattern: Points follow the regression line closely
- Conclusion: Strong predictor - larger enrollment consistently means more faculty
- Interpretation: Universities maintain relatively consistent student-faculty ratios regardless of size

**Summary:**

| Relationship | Correlation | Pattern | Predictive Power |
|-------------|-------------|---------|------------------|
| Tuition → Rank | Very Weak | Random | None |
| Acceptance → Graduation | Very Weak | Random | None |
| International % → Rank | Weak Negative | Slight trend | Low |
| Enrollment → Faculty | Strong Positive | Clear trend | High |

**Key Findings:**
- Only enrollment and faculty count show a strong, predictable relationship
- University rankings, tuition, acceptance rates, and graduation rates are influenced by many complex factors
- Simple linear relationships don't explain most university metrics
- The enrollment-faculty relationship makes logical sense: more students require more teachers

**Conclusion:**
Most university success metrics cannot be predicted by single variables. Only operational metrics like enrollment and faculty count show strong correlations, as they're directly related by institutional needs.
